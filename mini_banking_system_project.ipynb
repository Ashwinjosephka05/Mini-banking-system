```python
class BankAccount:
    def __init__(self, acc_number, pin, balance, name="", mobile=""):
        self.acc_number = acc_number
        self.pin = pin
        self.balance = balance
        self.name = name
        self.mobile = mobile
        self.transaction_history = []  # Initialize an empty list to store transaction history

    def deposit(self, amount):
        self.balance += amount
        self.transaction_history.append(f"Deposited {amount}")  # Record the deposit transaction
        return self.balance

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            self.transaction_history.append(f"Withdrew {amount}")  # Record the withdrawal transaction
            return self.balance

    def check_balance(self):
        return self.balance

    def get_transaction_history(self):
        return self.transaction_history

def change_pin(account):
    account_num=input("Enter the Account number:")
    new_name = input("Enter your username: ")
    new_mobile = input("Enter your registered mobile number: ")
    new_otp= input("Enter the otp:")
    new_pin = input("Enter new PIN:")
    account.pin = new_pin
    account.name = new_name
    account.mobile = new_mobile
    print("PIN changed successfully.")

def main():
    accounts = {
        "12345": {"pin": "1234", "balance": 0, "name": "ABC", "mobile": "1234567890"},
        "54321": {"pin": "4321", "balance": 0, "name": "XYZ", "mobile": "9876543210"},
    }

    while True:
        print("Welcome to our banking system")
        acc_number = input("Enter account number: ")
        if acc_number in accounts:
            pin = input("Enter PIN: ")
            if pin == accounts[acc_number]["pin"]:
                print("Login successful")
                current_account = BankAccount(
                    acc_number,
                    pin,
                    accounts[acc_number]["balance"],
                    accounts[acc_number]["name"],
                    accounts[acc_number]["mobile"],
                )
                break
            else:
                print("Incorrect PIN. Try again.")
        else:
            print("Account not found. Try again.")

    while True:
        print("\nPlease select an option:")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Change PIN")
        print("5. View Transaction History")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            amount = float(input("Enter amount to deposit: "))
            current_account.deposit(amount)
            print("Deposit successful. Current balance:", current_account.check_balance())
        elif choice == "2":
            amount = float(input("Enter amount to withdraw: "))
            if amount > current_account.check_balance():
                print("Insufficient funds")
            else:
                current_account.withdraw(amount)
                print("Withdrawal successful. Current balance:", current_account.check_balance())
        elif choice == "3":
            print("Current balance:", current_account.check_balance())
        elif choice == "4":
            change_pin(current_account)
        elif choice == "5":
            transaction_history = current_account.get_transaction_history()
            if transaction_history:
                print("Transaction History:")
                for transaction in transaction_history:
                    print(transaction)
            else:
                print("No transactions yet.")
        elif choice == "6":
            print("Exit.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
